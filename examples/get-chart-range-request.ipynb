```python
import json
with open("../credentials.json", "r") as f:
    CREDENTIALS = json.load(f)
```

```python
from datetime import datetime as dt
START = round(dt.today().replace(month=1, day=1).timestamp() * 1000)    # 1st January of the current year
END = round(dt.today().timestamp() * 1000)                              # Today (current date)
```

```python
import xapi
x = await xapi.connect(**CREDENTIALS)
result = await x.socket.getChartRangeRequest("PKN.PL_9", START, END, xapi.PeriodCode.PERIOD_D1, 0)
if result["status"] == True:
    rateInfos = result["returnData"]['rateInfos']
```

```python
import pandas as pd
df = pd.DataFrame(rateInfos)
df['open'] /= 1000
for column in ['close', 'high', 'low']:
    df[column] = df['open'] + df[column] / 1000
```
